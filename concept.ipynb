```python
from pyvis.network import Network
import pandas as pd 
import networkx as nx 
import random
```

```python
data = pd.read_csv('data.csv',sep=';')
```

```python
data
```

```python
generate_color = lambda : ["#"+''.join([random.choice('ABCDEF0123456789') for i in range(6)])] 
colors = {name: "#"+''.join([random.choice('ABCDEF0123456789') for i in range(6)]) for name in data['Inviters'].unique()}
colors_col = pd.DataFrame({colors[name] for name in data['Inviters']})
data['colors'] = colors_col


graph = nx.from_pandas_edgelist(data,source='Inviters',target='Name')
```

```python
net = Network(notebook=True, directed =True,height='750px', width='100%', bgcolor='#222222', font_color='white')
net.from_nx(graph)
```

```python
from cProfile import label


net.show_buttons(filter_=['physics'])
neighbor_map = net.get_adj_list()
names_list = data['Inviters'].unique()
generate_color = lambda : ["#"+''.join([random.choice('ABCDEF0123456789') for i in range(6)])] 
colors = {name: generate_color for name in names_list}
photos = dict(zip(data.Name,data.photo))
for node in net.nodes:
    if node['id'] =='Adam':
        node['color'] = 'yellow'
        node['size'] = 90
        node['shape'] = 'circularImage'
        node['image'] = r'images/papaj.jpg'
        
        
    else:
        try:
            if type(photos[node['id']]) == str:
                node['shape'] = 'circularImage'
                node['image'] = photos[node['id']]
        except:
            print("kupsko")

        node['value'] = len(neighbor_map[node["id"]])
        node['level'] = len(neighbor_map[node["id"]])
        
photos
```
